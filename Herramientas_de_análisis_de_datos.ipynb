```python
import numpy as np
import matplotlib.pyplot as plt


# Generar datos de ejemplo EN PYTHON
x = np.random.normal(size=100)  # 100 valores aleatorios de una distribución normal
y = x + np.random.normal(scale=0.5, size=100)  # y es x más un ruido aleatorio

# Crear un gráfico de dispersión
plt.scatter(x, y, color='blue')
plt.title("Gráfico de Dispersión")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Generar datos de ejemplo R
x <- rnorm(100)  # 100 valores aleatorios de una distribución normal
y <- x + rnorm(100, sd=0.5)  # y es x más un ruido aleatorio

# Crear un gráfico de dispersión
plot(x, y, main="Gráfico de Dispersión", xlab="X", ylab="Y", col="blue",
pch=19)
```
