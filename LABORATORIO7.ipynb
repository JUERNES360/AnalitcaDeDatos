```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('pvwatts_monthly.csv', on_bad_lines='skip') # This will skip lines that cause parsing errors
# Mostrar las primeras 5 filas del DataFrame
print(df.head())

# CREAR GRAFICA
# Generar datos de ejemplo EN PYTHON
x = np.random.normal(size=100)  # 100 valores aleatorios de una distribución normal
y = x + np.random.normal(scale=0.5, size=100)  # y es x más un ruido aleatorio

# Crear un gráfico de dispersión
plt.scatter(x, y, color='blue')
plt.title("Gráfico de Dispersion")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('pvwatts_monthly.csv', on_bad_lines='skip') # This will skip lines that cause parsing errors
# Mostrar las primeras 5 filas del DataFrame
print(df.head())

# Generar datos de ejemplo
x = np.arange(1, 11)  # Valores del 1 al 10
y = np.random.randint(1, 20, size=10)  # Valores aleatorios enteros

# Crear gráfico de barras
plt.bar(x, y)

# Títulos y etiquetas
plt.title("Gráfico de Barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")

# Mostrar gráfico
plt.show()
```
